# Discovery API

## Overview

The **Discovery API** enables searching, filtering, and sorting through a unified catalog of available Maxar content. It provides metadata information with selectable options about available content as **SpatioTemporal Asset Catalog (STAC)** and **GeoJSON**. Some result sets return links to order or stream the selected content.

## Key Parameters Used In Example

- **bbox** — `query`  
  Bounding box in format `"west,south,east,north"` in WGS84 decimal degrees.

- **collection** — `query`  
  Comma-separated list of collections to search in. If this parameter is not specified, items are retrieved from all available collections.

- **datetime** — `query`  
  Date range filter in format `"start-date/end-date"` or `"exact-datetime"`.

- **limit** — `query`  
  Maximum number of items to return.

- **sortby** — `query`  
  An array of property names, prefixed by either `+` for ascending or `-` for descending order.

- **area-based-calc** — `boolean`  
  Return area-based calculations in search results.

---

For more information, please refer to the [API Documentation](https://developers.maxar.com/docs/discovery/api/get-search-stac).


## Initial Discovery API Call

This following cell is responsible for retrieving geospatial data from the Maxar Discovery API. It performs the following tasks:

1. **User Input for API Key**:  
   The user is prompted to securely input their Maxar API key, which is required for authentication.

2. **Define API Request Details**:  
   A URL pointing to the Maxar Discovery API is defined. A set of query parameters specifies the search criteria, including:
   - The geographic area (bounding box).
   - The collection type (e.g., satellite imagery from a specific source).
   - A date range to filter results.
   - Additional options to limit results and refine the search.

3. **API Call**:  
   A GET request is sent to the Maxar Discovery API using the specified URL, headers, and parameters.

5. **Process and Print Results**:  
   - The API's response is converted into a JSON format for further processing.
   - A message is displayed to indicate the data retrieval is complete.
   - The total number of images returned by the API is printed for the user.

```python
import requests
import json
import pandas as pd
from getpass import getpass

maxar_api_key = getpass("Enter your API Key")


url = "https://api.maxar.com/discovery/v1/search"

payload={'bbox':'-105.11585405930997, 39.635307296586035, -104.80686358079436, 39.82013946676261', 'collection': 'wv01', 'datetime': '2023-11-06T00:00:00Z/2024-11-06T00:00:00Z', 'limit': 1000, 'area-based-calc': 'true', 'sortby': 'datetime'}
headers = {
  'Accept': 'application/geo+json',
  'MAXAR-API-KEY': maxar_api_key
}

response = requests.request("GET", url, headers=headers, params=payload)

response_json = response.json()

print("Finished recieving data from MGP Discovery, please move to the next cell.")
print(f"Number of images recieved: {response_json['numberReturned']}")
```

# Displaying Search Results on an Interactive Map

This cell will display an interactive map using `ipyleaflet` to visualize the results retrieved from the Discovery API and allow us to explore the geographical boundaries of the search results and examine the metadata associated with each area. It performs the following tasks:

1. **Initialize the Map**:  
   Creates a base map centered on a specific location with OpenStreetMap tiles.

2. **Load GeoJSON Features**:  
   Retrieves GeoJSON data and prepares features for display.

3. **Pagination**:  
   Divides features into pages and dynamically displays them on the map.

4. **Navigation Controls**:  
   Adds "Previous" and "Next" buttons to navigate through the paginated features.



```python
from ipyleaflet import Map, GeoJSON, basemaps, basemap_to_tiles
import ipywidgets as widgets
from IPython.display import display


m = Map(center=(39.87, -105.045), zoom=9)
tile_layer = basemap_to_tiles(basemaps.OpenStreetMap.Mapnik)
m.add_layer(tile_layer)


geo_json_data = response.json()
features = geo_json_data["features"]
paginated_features = []


items_per_page = min(len(features), 25)
total_pages = (len(features) + items_per_page - 1) // items_per_page
current_page = 0
current_image = 0

def display_image():
    
    geo_json_layer = GeoJSON(data=paginated_features[current_image], style={"color": "#BD00FF90", "opacity": 1, "weight": 2})
    m.layers = [tile_layer, geo_json_layer]

def change_page():
    global paginated_features
    start_idx = current_page * items_per_page
    end_idx = min(start_idx + items_per_page, len(features))
    
    paginated_features = features[start_idx:end_idx]

    


prev_page_button = widgets.Button(description="Previous Page")
next_page_button = widgets.Button(description="Next Page")
prev_image_button = widgets.Button(description="Previous Image")
next_image_button = widgets.Button(description="Next Image")
page_label = widgets.Label()

def on_prev_page_click(b):
    global current_page
    global current_image
    if current_page > 0:
        current_image = 0
        current_page -= 1
    else: 
        current_page = total_pages - 1
    update_display()

def on_next_page_click(b):
    global current_page
    global current_image
    if current_page < total_pages - 1:
        current_image = 0
        current_page += 1
    else:
        current_page = 0
    update_display()

def on_prev_image_click(b):
    global current_image
    if current_image > 0:
        current_image -= 1
        update_display()
    else:
        on_prev_page_click(b)
        current_image = min(len(paginated_features),items_per_page) - 1
        update_display()

def on_next_image_click(b):
    global current_image
    if current_image < min(len(paginated_features),items_per_page) - 1:
        current_image += 1
        update_display()
    else:
        on_next_page_click(b)

def update_display():
    change_page()
    display_image()
    page_label.value = f"Image {current_image + 1} of {min(len(paginated_features),items_per_page)} from Page {current_page + 1} of {total_pages}"

prev_page_button.on_click(on_prev_page_click)
next_page_button.on_click(on_next_page_click)
prev_image_button.on_click(on_prev_image_click)
next_image_button.on_click(on_next_image_click)


update_display()

display(m)
display(widgets.HBox([prev_page_button, prev_image_button, page_label, next_image_button, next_page_button]))
```

## Recieve Cloud Coverage Data

This cell retrieves cloud coverage data from the Maxar Discovery API response. It performs the following tasks:

1. **Prepare URLs**:  
   Extracts cloud coverage links from the API response and appends the API key as a query parameter.

3. **Process Responses**:  
   Collects successful responses containing cloud coverage data.

4. **Prints Results**:  
   Prints the total number of images that include cloud coverage data.


```python
import requests
from concurrent.futures import ThreadPoolExecutor


urls_to_call = []
for feature in response_json["features"]:
    if "assets" in feature and "cloud-cover" in feature["assets"] and "href" in feature["assets"]["cloud-cover"]:
        urls_to_call.append(feature["assets"]["cloud-cover"]["href"])

urls_with_api_key = [f"{url}?maxar_api_key={maxar_api_key}" for url in urls_to_call]

def fetch_url(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            return response.json()
        else:
            print(f"Failed to fetch: {url}, Status Code: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return None

detailed_responses = []
with ThreadPoolExecutor(max_workers=15) as executor:
    results = executor.map(fetch_url, urls_with_api_key)

detailed_responses = [result for result in results if result is not None]

print(f"Number of images that include cloud coverage: {len(detailed_responses)}")
```

## Display Cloud Coverage Area

This cell visualizes cloud coverage areas on an interactive map using `ipyleaflet`. It performs the following tasks:

1. **Initialize the Map**:  
   Creates a base map centered on a specific location with OpenStreetMap tiles.

2. **Paginate Data**:  
   Divides cloud coverage data into pages for better visualization.

3. **Display GeoJSON Data**:  
   Dynamically adds GeoJSON features (representing cloud coverage) to the map.

4. **Add Navigation Controls**:  
   Includes "Previous" and "Next" buttons to navigate through the paginated data.


```python
from ipyleaflet import Map, GeoJSON, basemaps, basemap_to_tiles
import ipywidgets as widgets
from IPython.display import display


m = Map(center=(39.7, -105.2), zoom=9)
tile_layer = basemap_to_tiles(basemaps.OpenStreetMap.Mapnik)
m.add_layer(tile_layer)


items_per_page = 10
total_pages = (len(detailed_responses) + items_per_page - 1) // items_per_page
current_page = 0


def display_page(page):
    start_idx = page * items_per_page
    end_idx = min(start_idx + items_per_page, len(detailed_responses))
    

    paginated_features = detailed_responses[start_idx:end_idx]
    

    paginated_geojson_data = {
        "type": "FeatureCollection",
        "features": paginated_features,
    }
    

    geo_json_layer = GeoJSON(data=paginated_geojson_data, style={"color": "#BD00FF90", "opacity": 1, "weight": 2})
    m.layers = [tile_layer, geo_json_layer]


prev_button = widgets.Button(description="Previous")
next_button = widgets.Button(description="Next")
page_label = widgets.Label()

def on_prev_button_click(b):
    global current_page
    if current_page > 0:
        current_page -= 1
        update_display()

def on_next_button_click(b):
    global current_page
    if current_page < total_pages - 1:
        current_page += 1
        update_display()

def update_display():
    display_page(current_page)
    page_label.value = f"Page {current_page + 1} of {total_pages}"

prev_button.on_click(on_prev_button_click)
next_button.on_click(on_next_button_click)

update_display()
display(m)
display(widgets.HBox([prev_button, page_label, next_button]))
```
